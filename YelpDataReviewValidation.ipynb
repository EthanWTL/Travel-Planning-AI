1 Hotel

```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
hotels = []

# Open the file and read it line by line
with open('Philadelphia_Hotels.jsonl', 'r') as file:
    for line in file:
        # Parse each line as a JSON object and append to the list
        json_obj = json.loads(line.strip())  # strip() removes any leading/trailing whitespace
        hotels.append(json_obj)
```

```python
df = pd.json_normalize(hotels)
df = df[['business_id','name','review_count']]
df
```

```python
hotel_reviews = []

# Open the file and read it line by line
with open('Philadelphia_Hotels_Reviews.jsonl', 'r') as file:
    for line in file:
        # Parse each line as a JSON object and append to the list
        json_obj = json.loads(line.strip())  # strip() removes any leading/trailing whitespace
        hotel_reviews.append(json_obj)
```

```python
df['real_review_count'] = df['business_id'].apply(lambda x: sum(1 for review in hotel_reviews if review['business_id'] == x))
df
```

```python
average_review_count_diff = (df['review_count'] - df['real_review_count']).mean()
average_review_count_diff
```
