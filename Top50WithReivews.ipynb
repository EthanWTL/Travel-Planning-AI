```python
import json
import pandas as pd
import numpy as np
```

```python
data = []
with open('Datasets/Restaurants_pruned.jsonl', 'r') as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)
df.head()
```

Stars

```python
df['star_categories'] = pd.cut(df['stars'], bins=[0, 1, 2, 3, 4, 5],
                               labels=['0', '1', '2', '3', '4'],
                               include_lowest=True)
```

```python
df.head()
```

Price

```python
#get unique prices
df['price'].unique()
```

```python
df.dropna(subset=['price'], inplace=True)
df['price'].unique()
```

```python
df['price_categories'] = df['price'].map({'$': 1, '$$': 2, '$$$': 3, '$$$$': 4})
df.head()
```

Review

```python
#get the distribution of review counts
df['review_count'].describe()
```

```python
df['review_count_categories'] = pd.cut(df['review_count'], bins=range(0, 6001, 100), labels=range(60))
df.head()
```

good for meal

```python
#drop record if good_for_meal is 'Unknown'
df.drop(df[df['good_for_meal'] == 'Unknown'].index, inplace=True)
#drop record if good_for_meal is '1'
df.drop(df[df['good_for_meal'] == '1'].index, inplace=True)
df.drop(df[df['good_for_meal'] == 1].index, inplace=True)
df.dropna(subset=['good_for_meal'], inplace=True)
df.drop(df[df['good_for_meal'] == ''].index, inplace=True)
df.drop(df[df['good_for_meal'] == 'nan'].index, inplace=True)
```

```python
mealSet = []
#for each word in good_for_meal, create a column
for meal in df['good_for_meal']:
    #split based on , 
    for word in meal.split(','):
        word = word.strip()
        #add the word into a data structure that won't save the duplicates
        mealSet.append(str(word))

print(np.unique(mealSet))
        
```

```python
df['good_for_meal_categories'] = -1
df.head()
```

```python
df['good_for_meal'].value_counts()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
for i in range(df.shape[0]):
    good_for_meal = df.iloc[i]['good_for_meal']
    for meal in good_for_meal.split(','):
        #to lowercase
        meal = meal.strip().lower()
        if meal == 'lunch':
            df.loc[i,'good_for_meal_categories'] = 1
            break
        if meal == 'breakfast':
            df.loc[i,'good_for_meal_categories'] = 0
        if meal == 'dinner':
            df.loc[i,'good_for_meal_categories'] = 2
```

```python
#count how many -1 we have for good for meal categories
df['good_for_meal_categories'].value_counts()
```

```python
#drop rows with -1 for good for meal categories
df.drop(df[df['good_for_meal_categories'] == -1].index, inplace=True)
df.reset_index(drop=True, inplace=True)
df['good_for_meal_categories'].value_counts()
```

Country of the food

```python
df['country_of_food'] = -1
df.head()
```

```python
for i in range(df.shape[0]):
    country_of_food = df.iloc[i]['categories']
    for country in country_of_food.split(','):
        #to lowercase
        country = country.strip().lower()
        if country == 'american (new)':
            df.loc[i,'country_of_food'] = 0
        if country == 'american (traditional)':
            df.loc[i,'country_of_food'] = 0
        if country == 'mediterranean':
            df.loc[i,'country_of_food'] = 1
        if country == 'chinese':
            df.loc[i,'country_of_food'] = 2
        if country == 'mexican':
            df.loc[i,'country_of_food'] = 3
        if country == 'italian':
            df.loc[i,'country_of_food'] = 4
        if country == 'thai':
            df.loc[i,'country_of_food'] = 5
        if country == 'japanese':
            df.loc[i,'country_of_food'] = 6
        if country == 'asian fusion':
            df.loc[i,'country_of_food'] = 7
```

```python
#cast star_categories to int
df['star_categories'] = df['star_categories'].astype('int64')
df['review_count_categories'] = df['review_count_categories'].astype('int64')
```

```python
# assume df is your DataFrame
columns_of_interest = ['star_categories', 'price_categories', 'review_count_categories','good_for_meal_categories','country_of_food']

# calculate the weights
weights = (df[columns_of_interest] ** 2).sum(axis=1) / (df[columns_of_interest] ** 2).sum(axis=1).sum()

# sample 100 rows
sampled_df = df.sample(n=100, weights=weights)
```
