### Create folders of reviews, each folder for a business

##### hotels

```python
hotel_reivews = []
with open('../Datasets/Hotels_useful_reviews.jsonl', 'r') as f:
    for line in f:
        hotel_reivews.append(json.loads(line))
```

```python
hotels = []
with open('../Datasets/Hotels_pruned.jsonl', 'r') as f:
    for line in f:
        hotels.append(json.loads(line))
```

```python
for hotel in hotels:
    current_business_id = hotel['business_id']
    current_business_name = hotel['name'].replace('/',' ')
    reviews = ''
    for review in hotel_reivews:
        if review['business_id'] == current_business_id:
            reviews += review['text']
            reviews += '\n'
            
    with open('../Datasets/Reviews/Hotels/' + current_business_name + '_' + current_business_id + '.txt', 'w') as f:
        f.write(reviews)
              
```
